```python
import matplotlib.pyplot as plt
import numpy as np
# Increase font sizes for better readability
plt.rcParams.update({
    'font.size': 20,
    'axes.titlesize': 20,
    'axes.labelsize': 20,
    'xtick.labelsize': 20,
    'ytick.labelsize': 20,
    'legend.fontsize': 20
})
# Set a professional color palette
colors = ['#4CAF50', '#03A9F4', '#FFC107', '#FF9800', '#F44336']
plt.style.use('seaborn-v0_8-talk')

# Data from 'Efficient+Market+Signal+Prediction...pdf' and 'sensors-21-07957.pdf'
models = ['TCN', '1D-CNN', 'LSTM', 'GRU', 'ARIMA']
accuracy = [72.4, 71.86, 63.8, 60.5, 55.2]

# Create the bar chart
plt.figure(figsize=(12, 7))
bars = plt.bar(models, accuracy, color=colors[:len(models)])

# Add labels and title
plt.ylabel('Prediction Accuracy (%)')
plt.ylim(50, 75) # Start y-axis at 50 to emphasize the difference

# Add data labels
plt.bar_label(bars, fmt='%.1f%%')

# Add gridlines and clean layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the figure
plt.savefig('temporal_learner_comparison.png')
print("Generated temporal_learner_comparison.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Increase font sizes for better readability
plt.rcParams.update({
    'font.size': 20,
    'axes.titlesize': 20,
    'axes.labelsize': 20,
    'xtick.labelsize': 20,
    'ytick.labelsize': 20,
    'legend.fontsize': 20
})
# Set a professional color palette
colors = ['#03A9F4', '#4CAF50', '#FFC107']
plt.style.use('seaborn-v0_8-talk')

# Data from 'entropy-25-00219.pdf'
# This paper compared 30 methods and ranked the sentiment tools.
# "Vader" and "Blob" (TextBlob) topped the ranking 7 times, FinBERT 4 times.
models = [ 'FinBERT','Vader', 'TextBlob']
top_rankings = [7,7,4]

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(models, top_rankings, color=colors)

# Add labels and title
plt.ylabel('Number of Top Rankings in Study')

# Add data labels
plt.bar_label(bars, fmt='%d')

# Add gridlines and clean layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the figure
plt.savefig('sentiment_encoder_comparison.png')
print("Generated sentiment_encoder_comparison.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Increase font sizes for better readability
plt.rcParams.update({
    'font.size': 20,
    'axes.titlesize': 20,
    'axes.labelsize': 20,
    'xtick.labelsize': 20,
    'ytick.labelsize': 20,
    'legend.fontsize': 20
})
# Set a professional color palette
colors = ['#4CAF50', '#F44336']
plt.style.use('seaborn-v0_8-talk')

# Data from '3453693.pdf' (ATTN-CNN-LSTM vs CNN-LSTM)
models = ['With Attention Fusion', 'Without Attention (Concat)']
loss = [0.0019, 0.0024]

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(models, loss, color=colors)

# Add labels and title
plt.ylabel('Model Loss (Lower is Better)')

# Add data labels
plt.bar_label(bars, fmt='%.4f')

# Add gridlines and clean layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the figure
plt.savefig('attention_fusion_comparison.png')
print("Generated attention_fusion_comparison.png")
```
